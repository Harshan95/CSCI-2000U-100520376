```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

labels = [r'USA (46)', r'CHN (38)', 
r'GBR (29)', r'RUS (24)',r'KOR (13)',r'GER (11)',r'FRA (11)'
          ,r'ITA (8)',r'HUN (8)',r'AUS  (7)',r'OTHER (107)']
sizes = [15.23,12.58,9.6,7.94,4.30,3.64,3.64,2.65,2.65,2.318,35.43]
colors = ['silver', 'red', 'green', 'blue','white','yellow','purple'
          ,'seagreen','coral','orange','black']
patches, texts = plt.pie(sizes, colors=colors, startangle=90)
plt.legend(patches, labels, loc="right")
plt.title("2012 Olympics gold medals 100520376")
# Set aspect ratio to be equal so that pie is drawn as a circle.
plt.axis('equal')
plt.tight_layout()
plt.show()
```
