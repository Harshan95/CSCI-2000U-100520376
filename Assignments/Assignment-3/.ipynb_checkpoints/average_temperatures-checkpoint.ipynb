```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import scipy

f = open('toronto_temperature_2012.dat')
line = f.readline()
toronto = [None] * 13;
counter = 0;
while line:
    toronto[counter+1]=float(line)
    ## reads each line because each line contains a number
    line = f.readline()
    counter = counter + 1;
f.close() 

f = open('montreal_temperature_2012.dat')
line = f.readline()
montreal = [None] * 13;
counter = 0;
while line:
    montreal[counter+1]=float(line)
    ## reads each line because each line contains a number
    line = f.readline()
    counter = counter + 1;
f.close() 

f = open('vancouver_temperature_2012.dat')
line = f.readline()
vancouver = [None] * 13;
counter = 0;
while line:
    vancouver[counter+1]=float(line)
    ## reads each line because each line contains a number
    line = f.readline()
    counter = counter + 1;
f.close() 

fig, ax = plt.subplots()
plt.plot (toronto, color='blue', alpha=1.00)
plt.plot (montreal, color='red', alpha=1.00)
plt.plot (vancouver, color='green', alpha=1.00)
labels = [r'Toronto',r'Montreal',r'Vancouver']
plt.legend(labels, loc="best")
plt.ylabel('Temperature')
plt.xlabel('Months')

fig.canvas.draw()

labels = [item.get_text() for item in ax.get_xticklabels()]
labels[1] = 'Feb'
labels[2] = 'Apr'
labels[3] = 'June'
labels[4] = 'Aug'
labels[5] = 'Oct'
labels[6] = 'Dec'

plt.title("2012 Average Monthly Temperatures 100520376")
ax.set_xticklabels(labels)                                                    

plt.show()
```
